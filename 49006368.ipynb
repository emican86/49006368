```python
import pandas as pd
import pandas_datareader as pdr
```

```python
from datetime import datetime
```

```python
start = datetime(2018, 2, 1)
end = datetime(2018,2,26)
googDF = pdr.data.DataReader('goog', 'morningstar', start,end)
```

```python
import talib
```

```python
googDF
```

```python
talib.STOCH?
```

```python
googDF['slowk'], googDF['slowd'] = talib.STOCH(
                                                high=googDF.High.as_matrix()
                                              , low=googDF.Low.as_matrix()
                                              , close=googDF.Close.as_matrix()
                                              , fastk_period=5
                                              , slowk_period=3
                                              , slowk_matype=0
                                              , slowd_period=3
                                              , slowd_matype=0)
```

#### It appears the default lookback is set to 8 days.  To get the slowk and slowd indicators for February, we must go 8 days prior

```python

start = datetime(2018, 1, 19)
end = datetime(2018,2,26)
googDF2 = pdr.data.DataReader('goog', 'morningstar', start,end)
```

```python
googDF2['slowk'], googDF2['slowd'] = talib.STOCH(
                                                  high=googDF2.High.as_matrix()
                                                , low=googDF2.Low.as_matrix()
                                                , close=googDF2.Close.as_matrix()
                                                , fastk_period=5
                                                , slowk_period=3
                                                , slowk_matype=0
                                                , slowd_period=3
                                                , slowd_matype=0)
```
